```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_data = pd.read_csv("train/train.csv")
train_labels = pd.read_csv("train_labels.csv")
```

```python
df = train_data.merge(train_labels, how="left", on="sequence")
```

```python
df.describe()
```

```python
sequence_x = df[df.sequence == 100]
plt.plot(sequence_x.step, sequence_x.sensor_00)
plt.show()
```

```python
df_full_corr = sequence_x[[f for f in sequence_x.columns if f.startswith("sensor_")]].corr()
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(df_full_corr, annot=True)
```
