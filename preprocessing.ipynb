```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras as ks
import tensorflow as tf
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
train_data = pd.read_csv("data/train.csv")
train_labels = pd.read_csv("data/train_labels.csv")
```

```python
train_data=train_data.drop(columns=["subject"])
```

```python
train_vals = train_data.values
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
```

```python
scaler.fit(train_vals)
```

```python
train_data=pd.DataFrame(scaler.transform(train_vals),columns=train_data.columns)
```

```python
train_data
```

```python
pivot_df = train_data.pivot(index="sequence", columns="step", values=[f for f in train_data.columns if f.startswith("sensor_")])
```

```python
train_labels=train_labels.drop(columns="sequence")
```

```python
pivot_df
```

```python
train_labels
```

```python
train_vals = pivot_df.values
```

```python
train_labels_vals = train_labels.values
```

```python
TRAIN_LIMIT = 0.8
```

```python
train_size = int(len(train_labels_vals) * TRAIN_LIMIT)
trainX, testX = train_vals[0:train_size,:], train_vals[train_size:,:]
trainY, testY = train_labels_vals[0:train_size,:], train_labels_vals[train_size:,:]
```

```python
trainX.shape, trainY.shape
```

```python
testX.shape, testY.shape
```

```python
trainX = trainX.reshape(trainX.shape[0], trainX.shape[1], 1)
testX = testX.reshape(testX.shape[0], testX.shape[1], 1)
```

```python
trainY = trainY.reshape(trainY.shape[0])
testY = testY.reshape(testX.shape[0])
```

```python
testX.shape, testY.shape
```

```python
trainX.shape, trainY.shape
```

```python
model = ks.Sequential([
    ks.layers.LSTM(32, return_sequences=True, activation='relu'),
    ks.layers.Dropout(.2),
    ks.layers.LSTM(32, activation='relu'),
    ks.layers.Dropout(.2),
    # ks.layers.Flatten(),
    ks.layers.Dense(1, activation="sigmoid"),
])
```

```python
initial_learning_rate= 0.001
```

```python
model.compile(
    loss=ks.losses.BinaryCrossentropy(),
    # optimizer=tf.optimizers.Adam(learning_rate=0.01, epsilon=1e-9, beta_1=0.9, beta_2=0.98),
    optimizer=tf.optimizers.Adam(learning_rate=initial_learning_rate),
    metrics=[ks.metrics.AUC(), ks.metrics.Accuracy()],
)
```

```python
epochs=20
```

```python
trainX.shape, testX.shape
```

```python
trainY.shape, testY.shape
```

```python
history = model.fit(
    trainX,
    trainY,
    epochs=epochs,
    verbose=2,
    shuffle=True,
    # callbacks=[
    #     early_stopping,
    #     lr_scheduler,
    #     csvlogger,
    #     checkpoints
    # ],
    validation_data=[testX, testY],
    batch_size=256
)
```
